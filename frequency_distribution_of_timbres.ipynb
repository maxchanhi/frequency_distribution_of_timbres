```python
import pandas as pd
```

```python
path = "/content/drive/MyDrive/Research data/5 ET temperament calculator - ratio.csv"
df=pd.read_csv(path)
#df = pd.DataFrame()
data_list = df[df.columns[0]].values.tolist()

range = [1,3.18]
d = 12*6
celling_gp = (range[1]-range[0])/d
gp_range = []
gp_range_mid_pt = []

i = 0
while i < d:
  class_celling = (celling_gp)*(i+1)+1
  class_floor= (celling_gp)*i+1
  gp_range.append(class_floor)
  #gp_range.append(class_celling)
  i += 1
# Calculate the midpoint for each interval
i = 0
while i < len(gp_range) - 1:
    lower_bound = gp_range[i]
    upper_bound = gp_range[i + 1]
    midpoint = (lower_bound + upper_bound) / 2
    gp_range_mid_pt.append(midpoint)
    i += 1
print(gp_range_mid_pt)
print(gp_range)
fit_count = []
tally_marks = []
i = 0
g = 0
while g < len(gp_range)-1:
  while i < len(data_list):      
    if data_list[i] > gp_range[g] and data_list[i] <= gp_range[g+1]:
      tally_marks.append("T")
    else:
      pass
    i += 1
  fit_count.append(len(tally_marks))
  tally_marks  = []
  g += 1
  i = 0
print(fit_count)
```

```python
i=1
modes = []
while i < len(fit_count)-1:
  cap_l= gp_range[i]
  l = fit_count[i]-fit_count[i-1]
  u = fit_count[i]-fit_count[i+1]
  c = (range[1]-range[0])/d
  if l+u == 0:
      modes.append(0)
      i += 1  
  else:
      modes.append(cap_l+(l/(l+u)*c))
      i += 1
print(modes)
```

```python
from collections import Counter

# data_list is a list of values
# gp_range is a list of upper and lower bounds for each class
# fit_count is a list of counts for each class

# Initialize an empty list to store the modes
modes = []
i=0
# Iterate over each class
for i in fit_count:
    # Get the upper and lower bounds of the current class
    lower_bound = gp_range[i]
    upper_bound = gp_range[i + 1]
    
    # Get the count for the current class
    count = fit_count[i]
    
    # If the count is 0, the mode is None
    if count == 0:
        modes.append(None)
        continue
    
    # Otherwise, get the values in the current class
    values = [val for val in data_list if lower_bound <= val < upper_bound]
    
    # Use the Counter class to count the number of occurrences of each value
    value_counts = Counter(values)
    
    # Get the value with the highest count
    mode = value_counts.most_common(1)[0][0]
    
    # Store the mode in the modes list
    modes.append(mode)
    i+=1
# modes is a list of the modes for each class
print(modes)
```
